### Percentage 

```python
# Show Histogram of the percentage

mu = first['Percentage'].mean() # mean of distribution
std = first['Percentage'].std() # standard deviation of distribution
data = first['Percentage']

mu, std = norm.fit(data) 
  
# Plot the histogram.
plt.hist(data, bins=25, density=True, alpha=0.6, color='b')
  
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
  
plt.plot(x, p, 'k', linewidth=2)
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)
  
plt.show()
```

## Percentage for science branch:

```python
# collect the data for each governorate 
first_science = first.query('branch == "Science"')
percent_gove_science = first_science.pivot(index='desk_no', columns='Governorate', values='Percentage')


#MultipleBoxplot

percent_gove_science.boxplot(figsize=(20, 7))
plt.xticks(rotation=90)
plt.show()
```

```python
# collect the data for each region 
percent_reg_science = first_science.pivot(index='desk_no', columns='Region', values='Percentage')


#MultipleBoxplot

percent_reg_science.boxplot(figsize=(20, 10))
plt.xticks(rotation=90)
plt.show()
```

#### Conclusion:
>- Percentage data are normally distrubuted, and litte skewed to left (to the average of the percent).
>- On governorate level: governorates with highest percentage are Dakahlia and North Sinim while Giza and Menofia are the lowest.
>- One regional level: Delta and Suize Canal regions have the highest percentage, while Gereator Cairo and Central Uper Egypt are the lowest.
