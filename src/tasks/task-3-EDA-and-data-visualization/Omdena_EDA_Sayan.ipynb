```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"/content/SecondExam_Results_with_gender_soliman.csv", index_col=0)
df.head(2)
```

```python
df = df.replace(['unscheduled','absent','canceled','إجتياز'], 0)
df
```

```python
df['student_type'].value_counts()
```

## Top 15 states from where most students appear for the exam

```python
value=df['government'].value_counts().head(15).values.tolist()
index =df['government'].value_counts().head(15).index.tolist()

plt.figure(figsize=(18,10))
palette_color = sns.color_palette('bright')
plt.pie(value, labels=index, colors=palette_color,autopct='%1.1f%%')
  
plt.show()
```

## Distribution of Gender across different streams

```python


plt.figure(figsize=(17,6))
ax = sns.countplot(x ='branch', hue = "gender", data = df, order=df['branch'].value_counts().index)
plt.xticks(rotation=90)

plt.show()
```

```python
subjects = df[['government','gender','administration','pure_mathematics', 'history', 'geography', 'philosophy', 'psychology','chemistry', 'biology', 'geology', 'applied_math', 'physics']]
subjects
```

```python
subjects[['pure_mathematics', 'history', 'geography', 'philosophy', 'psychology','chemistry', 'biology', 'geology', 'applied_math', 'physics']] = subjects[['pure_mathematics', 'history', 'geography', 'philosophy', 'psychology','chemistry', 'biology', 'geology', 'applied_math', 'physics']].astype(float).astype(int)
```

```python
subjects['history'].value_counts().index
```

```python
subjects.plot(kind='box')
plt.xticks(rotation=90)
plt.show()
```

## BoxPlot of Subjects w.r.t different governments

```python
names = ['pure_mathematics', 'history', 'geography', 'philosophy', 'psychology','chemistry', 'biology', 'geology', 'applied_math', 'physics']

ncols = len(names)
fig, axes = plt.subplots(5,2,sharex=False,sharey=False,figsize=(30,40))
#plt.figure(figsize=(20,6)) #not mandatory

for name, ax in zip(names, axes.flatten()):
    axa = sns.boxplot(y=name, x= "government", data=subjects, orient='v', ax=ax)
    for tick in axa.get_xticklabels():
        tick.set_rotation(90)
plt.tight_layout()
```

## Boxplot of Different subjects w.r.t Gender

```python
names = ['pure_mathematics', 'history', 'geography', 'philosophy', 'psychology','chemistry', 'biology', 'geology', 'applied_math', 'physics']

ncols = len(names)
fig, axes = plt.subplots(5,2,sharex=False,sharey=False,figsize=(30,40))
plt.figure(figsize=(20,6))

for name, ax in zip(names, axes.flatten()):
    axa = sns.boxplot(y=name, x= "gender", data=subjects, orient='v', ax=ax)
    for tick in axa.get_xticklabels():
        tick.set_rotation(90)
plt.tight_layout()
```
