## Concatenating the datasets and updating values

```python
#updating values in the first dataset
df = pd.concat([first,second]).drop_duplicates(subset=["desk_no","name"],keep="last")
```

```python
df.index =np.arange(0,len(df))
```

```python
cols = ["government_English","branch","student_type","status"]
```

## Data cleaning

```python
#dropping redundant columns
df = df.drop(["student_type","schoolName_Arabic","administration_Arabic","government_Arabic"],axis=1)
```

```python
inconsistency = ["مؤجل","معفى","إجتياز","unscheduled","absent","canceled"]
```

```python
#replacing inconsistencies with nan values
df = df.replace(inconsistency,np.nan)
```

```python
#converting columns with scores to float datatype
subjects = ["first_foreign_lang","second_foreign_lang","pure_mathematics","history","geography","philosophy","psychology"
           ,"chemistry","biology","geology","applied_math","physics","total"]
for subject in subjects:
    df[subject]=df[subject].astype("float64")
```

```python
#converting percentage to float
df["Percentage"] = df["Percentage"].apply(lambda percentage: float(percentage[:-1]))
```

```python
#fixing typo
df["status"] = df["status"].replace("second Trail","second trial")
```

## Data visualization

```python
df["government_English"].value_counts().sort_values(ascending=False).plot(kind="bar")
plt.title("Region wise attempts")
plt.show()
```

```python
plt.figure(dpi=125)
sns.boxplot(data=df,x="government_English",y="Percentage")
plt.xticks(rotation=90)
plt.title("Region vs Student performance")
plt.show()
```

```python
print("Top performing regions")
df[["government_English","Percentage"]].groupby("government_English").median().sort_values(by="Percentage",ascending=False).plot(kind="bar");
```

```python
df[df["status"]=="successful"]["government_English"].value_counts(ascending=False).plot(kind="bar")
plt.title("Successful attempts")
plt.show()
```

```python
df[df["status"]=="fail"]["government_English"].value_counts(ascending=False).plot(kind="bar")
plt.title("Failed attempts")
plt.show()
```

### Observation
* Maximum attempts were made from Cairo, Giza, Sharqia, Alexandria and Dakahlia.
* It appears that the top 5 performing regions are North Sinai, Dakahlia, Port Said, Damietta and South Sinai.
* Least performing regions are Luxor, Aswan, Asyut, Giza and Minya.
* Maximum successful attempts were made from Cairo, Giza and Sharqia.
* Maximum failed attempts were from Giza, Cairo and Minya.

```python
sns.countplot(data=df,x="status",order=["successful","fail"],hue="gender")
plt.title("Status of students w.r.t gender")
plt.show()
```

```python
scores = df[subjects+["gender"]]
for col in scores.columns[:-1]:
    x=scores[col][~np.isnan(scores[col])]
    sns.kdeplot(data=scores,x=x,hue="gender",fill=True);
    plt.title("Subject wise scores of both genders")
    plt.show()
```

### Observation
* Females have higher scores than males in all subjects except pure mathematics and applied mathematics.
